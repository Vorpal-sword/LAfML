Exercise
Jill designs solar panels as a hobby.

On April 1st, Jill’s “Mark I” design begins generating power: 1 kJ/day.
On May 1st, her “Mark II” design begins generating 4 kJ of power per day.

1. What day is it when Jill’s Mark II design has generated as much total energy as the Mark I design?

2. How much total energy have both generated by that day?

3. What would the solutions to (1.) and (2.) be if Mark II design generated 1 kJ of power per day?


```python
# just import libs
import numpy as np # for numbers almost always
import matplotlib.pyplot as plt # for plottind
```

```python
t = np.linspace(0, 13, 1000) # start, stop, n points

# variables
markOne = 1*(t+30)
markTwo = 4*t
```

```python
fig, ax = plt.subplots()
plt.title('Energy by days')
plt.xlabel('Days')
plt.ylabel('Energy')
ax.set_xlim([0, 13])
ax.set_ylim([0, 50])
ax.plot(t, markOne, c='green')
ax.plot(t, markTwo, c='brown')
plt.axvline(x= 10, color='purple', linestyle='--')
_ = plt.axhline(y= 40, color='purple', linestyle='--')

# I wanna add for green and brown marks description(legend)

# Add labels to the plots
ax.plot(t, markOne, c='green', label='Mark I (1 kJ/day)')
ax.plot(t, markTwo, c='brown', label='Mark II (4 kJ/day)')

# Show legend
plt.legend()

plt.show() #works without it?

#In Jupyter Notebooks (or other interactive environments like IPython), plt.show() is not strictly required — plots often display automatically.

#In scripts or regular Python files, you should include plt.show() at the end to explicitly display the plot window. Without it, the plot may not appear at all.
```

Answers:

1. 10 days
2. 80kJ
3. NO solutions
